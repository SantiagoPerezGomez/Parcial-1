```python
#Punto 2
def factorial(num): 
    if num < 0: 
        return 0

    elif num == 0: 
        return 1
        
    else: 
        fact = 1
        while(num > 1): 
            fact *= num 
            num -= 1
        return fact 
    
def conbinatoria(m, n):
    result_m = factorial(m)
    result_n = factorial(n)
    result_mn = factorial(m-n)
    if result_m == 0 and result_n == 0:
        print("El factorial de un numero negativo no existe")
    else:
        return result_m / (result_n*(result_mn))
    
    
def run():
    m = int(input("Ingrese el valor de M = "))
    n = int(input("Ingrese el valor de N = "))
    
    print(f'la combinatoria de {m} y {n} es : {conbinatoria(m, n)}')
    
    
if __name__ == '__main__':
    run()
```

```python
#Punto 3
def sumatoria(opcion, valor): 
    if opcion == 1: 
        print(f"Σ de i=1 hasta {valor} donde i(i+1)")
        i = 1
        suma = 0
        while i<=valor:
            suma += i*(i+1)
            i+=1
        return suma

    elif opcion == 2: 
        print(f"Σ de i=0 hasta {valor} donde 2^n")    
        i = 0
        suma = 0
        while i<=valor:
            suma += 2**i
            i+=1
        return suma
    else: 
        print(f"Σ i=1 hasta {valor} donde 4*n")    
        i = 1
        suma = 0
        while i<=valor:
            suma += 4*i
            i+=1
        return suma
    
    
def run():
    print("Escoja la sumatoria a resolver:")
    print("1) Σ de i=1 hasta n donde i(i+1)")
    print("2) Σ de i=0 hasta n donde 2^n")    
    print("3) Σ i=1 hasta n donde 4*n")   
    opcion = int(input("Ingrese una opcion:"))
    if  opcion > 0 and opcion < 4:
        valor = int(input("Ingrese un valor para n = "))
        print(f"Respuesta: {sumatoria(opcion, valor)}")
    else:
        print("su opcion no es valida")
    
    
    
    
    
if __name__ == '__main__':
    run()
```
